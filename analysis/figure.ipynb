```python
import pandas as pd
from astropy.table import Table, QTable
import astropy.units as u
from astropy.time import Time
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cfht_eqw_results.csv')
data.rename(columns={'Time(Y,M,D,H,M,S)':'time'}, inplace=True)
data['BJD'] = np.zeros(len(data))
bjds = []
files = []
for obj in data.itertuples():
    filename = obj[1]
    year = obj[3].split()[0]
    y = int(year.strip('(,'))
    month = obj[3].split()[1]
    m = int(month.strip(','))
    day = obj[3].split()[2]
    d = int(day.strip(','))
    hour = obj[3].split()[3]
    h = int(hour.strip(','))
    minute = obj[3].split()[4]
    min = int(minute.strip(','))
    second = obj[3].split()[5]
    s = int(second.strip(',)'))
   
    bjd = Time(datetime(y,m,d,h,min,s), scale='utc')

    bjds.append(bjd)
    files.append(filename)

data['BJD'] = bjds
#data
```

```python
times_a= np.array(data['BJD'])
times = [t.to_value('mjd') for t in times_a]
widths = np.array(data['EQW_Value'])

plt.figure(figsize=(12, 6))
plt.scatter(times, widths, s=25)
plt.xlabel('Time (BJD)')
plt.ylabel('H$\u03B1$ Equivalent Width')
#plt.title('')
plt.ylim(0,0.6)
plt.title('Full Observational Timescale')
plt.show()
```

```python
t0 = times[0]
period = 3.5247 #days, from Exoplanet Archive
phase = ((times-t0) % period) #known_period 

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(phase, widths, s=25)
ax.set_xlabel('Phase')
ax.set_ylabel('H$\u03B1$ Equivalent Width')
ax.set_ylim(0,0.6)
ax.set_title('Phase-Folded')
```
