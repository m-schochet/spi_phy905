```python
import pandas as pd
from astropy.table import Table, QTable
import astropy.units as u
from astropy.time import Time
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cfht_eqw_results.csv')
data.rename(columns={'Time(Y,M,D,H,M,S)':'time'}, inplace=True)
data['MJD'] = np.zeros(len(data))
mjds = []
files = []
for obj in data.itertuples():
    filename = obj[1]
    year = obj[3].split()[0]
    y = int(year.strip('(,'))
    month = obj[3].split()[1]
    m = int(month.strip(','))
    day = obj[3].split()[2]
    d = int(day.strip(','))
    hour = obj[3].split()[3]
    h = int(hour.strip(','))
    minute = obj[3].split()[4]
    min = int(minute.strip(','))
    second = obj[3].split()[5]
    s = int(second.strip(',)'))
   
    mjd = Time(datetime(y,m,d,h,min,s), scale='utc').to_value('mjd')

    mjds.append(mjd)
    files.append(filename)

data['MJD'] = mjds
#data
```

```python
times= np.array(data['MJD'])
widths = np.array(data['EQW_Value'])

times2= np.array(df['MJD'])
widths2 = np.array(df['EQW_Value'])

fig, ax = plt.subplots(figsize=(12, 6))#.figure(figsize=(12, 6))
ax.scatter(times, widths, s=25, label='CFHT/ESPaDOnS')
ax.scatter(times2, widths2, s=25, label='SOAR/GOODMAN')
ax.legend()
ax.set_xlabel('Time (BJD)')
ax.set_ylabel('H$\u03B1$ Equivalent Width')
#ax.set_ylim(0,1.0)
ax.set_title('Full Observational Timescale')
```

```python
# phase fold to planet period
period = 3.5247 # days
phases = (times - times[0]) % period / period
phases2 = (times2 - times[0]) % period / period
plt.figure(figsize=(12, 3))
plt.scatter(phases, widths, s=25, label='CFHT/ESPaDOnS')
plt.scatter(phases2, widths2, s=25, label='SOAR/GOODMAN')
plt.legend()
plt.xlabel('Phase (Period = 3.5247 days)')
plt.ylabel('H$\u03B1$ Equivalent Width')
plt.title('Phase Folded H$\u03B1$ Equivalent Widths')
plt.grid(True)
plt.show()
```

```python
# curve fit to data

all_phases = np.concatenate((phases, phases2))
all_widths = np.concatenate((widths, widths2))

# filter out nans
valid_indices = ~np.isnan(all_widths)
all_phases = all_phases[valid_indices]
all_widths = all_widths[valid_indices]

from scipy.optimize import curve_fit
def sine_func(x, a, c, d):
    return a * np.sin(2 * np.pi * x + c) + d
popt, pcov = curve_fit(sine_func, all_phases, all_widths, p0=[0.1, 0, 0.9])
# generate fit line
x_fit = np.linspace(0, 1, 100)
y_fit = sine_func(x_fit, *popt)
plt.figure(figsize=(12, 3))
plt.scatter(phases, widths, s=25, label='CFHT/ESPaDOnS')
plt.scatter(phases2, widths2, s=25, label='SOAR/GOODMAN')
plt.plot(x_fit, y_fit, color='red', label='Sine Fit')
plt.legend()
plt.xlabel('Phase (Period = 3.5247 days)')
plt.ylabel('H$\u03B1$ Equivalent Width')
plt.title('Phase Folded H$\u03B1$ Equivalent Widths with Sine Fit')
plt.grid(True)
plt.show()

# replot with subplots, residuals plot below
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), gridspec_kw={'height_ratios': [3, 1]}, sharex=True)
ax1.scatter(phases, widths, s=25, label='CFHT/ESPaDOnS')
ax1.scatter(phases2, widths2, s=25, label='SOAR/GOODMAN')
ax1.plot(x_fit, y_fit, color='red', label='Sine Fit')
ax1.legend()
ax1.set_ylabel('H$\u03B1$ Equivalent Width')
ax1.set_title('Phase Folded H$\u03B1$ Equivalent Widths with Sine Fit')
ax1.grid(True)
# calculate residuals
residuals = all_widths - sine_func(all_phases, *popt)
ax2.scatter(all_phases, residuals, s=25, color='orange', label='Residuals')
ax2.axhline(0, color='red', linestyle='--')
ax2.set_xlabel('Phase (Period = 3.5247 days)')
ax2.set_ylabel('Residuals')
ax2.legend()
ax2.grid(True)
plt.show()

# use covariance to estimate uncertainties
perr = np.sqrt(np.diag(pcov))
print("Fitted parameters:")
print(f"a (amplitude): {popt[0]:.4f}  {perr[0]:.4f}")
print(f"c (phase shift): {popt[1]:.4f}  {perr[1]:.4f}")
print(f"d (vertical shift): {popt[2]:.4f}  {perr[2]:.4f}")

# calculate r squared
residuals_all = all_widths - sine_func(all_phases, *popt)
ss_res = np.sum(residuals_all**2)
ss_tot = np.sum((all_widths - np.mean(all_widths))**2)
r_squared = 1 - (ss_res / ss_tot)
print(f"R-squared: {r_squared:.4f}")
```
